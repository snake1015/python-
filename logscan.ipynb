```python
'''
 作者：snake
 描述：
     解压所有位于指定目录下的日志ZIP压缩文件
     扫描日志分析所有日志中的通话标识：connid
     将所有扫描出来的connid写入文本connidlist.txt
 日期： 12/16/2017
 
'''
import os
import re
import sys
import zipfile

class logscan:
    def __init__(self,filepath):
        self.filepath = filepath
        
    def getfilelist(self):
        '''获取目录下的所有文件名'''
        files = os.listdir(self.filepath)
        print(files)
        return(files)
    
    def readfile(self,path):
        '''读取文件内容'''
        fr = open(path)
        filetext = []
        try:
            for line in fr.readlines():
                #print(line)
                filetext.append(line)
        finally:
            fr.close()
        return(filetext)
    
    def writefile(self,path,content):
        '''将内容写入文件中'''
        fw = open(path,'a')
        try: 
            if isinstance(content,(list)):
                for i in content:
                    fw.write(i)
                    #fw.write('\n')
            else:
                fw.write(content)
        finally:
            fw.close()
    
    def getallconnid(self,path):
        '''获取日志文件中所有的connid,并去重'''
        filetext = self.readfile(path)
        connidlisttmp = []
        for line in filetext:
            if re.search('AttributeConnID',line):
                connidlisttmp.append(line.split('\t')[2])
            else:
                continue
        connidlist = list(set(connidlisttmp))
        print(connidlist)
        savepath = os.path.dirname(path)+ '\connidlist.txt'
        print(savepath)
        print(connidlisttmp)
        self.writefile(savepath,connidlist)
        
        
    def extractfile(self):
        '''解压所有压缩文件到当前目录'''
        for file in self.getfilelist():
            fullpath = os.path.join(path,file)
            if zipfile.is_zipfile(fullpath):
                zfile = zipfile.ZipFile(fullpath,'r')
                for i in zfile.infolist():
                    zfile.extract(i,path=os.path.dirname(fullpath))
                #print(zfile.readlines(i.filename))

if __name__ == '__main__':
    path = input('the log file path: ' )
    fp = logscan(path)
    fp.extractfile()
    for file in fp.getfilelist():
        fullpath = os.path.join(path,file)
        if not zipfile.is_zipfile(fullpath):
            fp.readfile(fullpath)
            fp.getallconnid(fullpath)
    
        
    


```
